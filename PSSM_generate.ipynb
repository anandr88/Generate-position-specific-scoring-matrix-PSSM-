### possum.py usage:
##### python possum.py <options> <source files>
##### -i,--input: input a file in fasta format.
##### -o,--output: output a file of the generated feature.
##### -t,--type: specify a feature encoding algorithm.
    example: aac_pssm, d_fpssm, smoothed_pssm, ab_pssm, pssm_composition, rpm_pssm, s_fpssm, dpc_pssm
             k_separated_bigrams_pssm, tri_gram_pssm, eedp, tpc, edp, rpssm, pse_pssm, dp_pssm, pssm_ac
             pssm_cc, aadp_pssm, aatp, medp
##### -p,--pssmdir: specify the directory of PSSM files.    
    
            #### Row Transformation-Based Features
    
            These methods operate on the rows of the PSSM matrix, which correspond to positions in the sequence.
            1.	AAC-PSSM (Amino Acid Composition from PSSM):
            o	Calculates the average activation values for each amino acid position across the sequence.
            o	Captures how strongly each amino acid type is represented.
    
            2.	D-FPSSM (Difference-based FPSSM):
            o	Computes a normalized difference between rows to emphasize variations in PSSM values.
            o	Highlights position-specific evolutionary information.
    
            3.	Smoothed-PSSM:
            o	Applies a smoothing window across rows to reduce noise.
            o	Ensures positional continuity in evolutionary scores.
    
            4.	AB-PSSM (Average Block PSSM):
            o	Divides the PSSM matrix into 20 equal blocks and calculates AAC-PSSM for each block.
            o	Aggregates information from different regions of the protein.
    
            5.	PSSM Composition:
            o	Treats the PSSM matrix as a single vector and computes its average.
            o	Encodes global properties of the matrix.
    
            6.	RPM-PSSM (Relative Position Matrix):
            o	Captures position-relative differences in the PSSM.
            o	Useful for identifying regions of significant evolutionary conservation.
    
            7.	S-FPSSM (Sequence-based FPSSM):
            o	Converts the entire PSSM into a long feature vector.
            o	Encodes sequential and evolutionary information.

            #### Column Transformation-Based Features
            These methods operate on the columns of the PSSM matrix, which correspond to amino acid substitution scores.
            1.	DPC-PSSM (Di-Peptide Composition from PSSM):
            o	Calculates the frequency of di-peptide pairs based on substitution scores.
            o	Captures dependencies between consecutive residues.
    
            2.	K-Separated Bigrams PSSM:
            o	Similar to DPC-PSSM but considers residues separated by k positions.
            o	Useful for identifying long-range interactions.
    
            3.	Tri-Gram PSSM:
            o	Encodes triplets of residues for detailed local interaction patterns.
            o	Generates higher-dimensional features.
    
            4.	EEDP (Evolutionary Difference PSSM):
            o	Computes the evolutionary difference between PSSM values for paired positions.
            o	Enhances the distinction between conserved and variable regions.
    
            5.	TPC (Triplet Position Composition):
            o	Calculates residue-level differences and normalizes substitution probabilities.
            o	Encodes sequence-level variations across triplets.

            #### Mixed Row and Column Features
            These methods combine row-wise and column-wise transformations.
            1.	EDP (Evolutionary Difference Profile):
            o	Computes differences between PSSM values for paired residues across rows and columns.
            o	Highlights evolutionary variation.
    
            2.	RPSSM (Reverse PSSM):
            o	Encodes reverse-order information from the PSSM.
            o	Useful for capturing complementary patterns.
    
            3.	PSE-PSSM (Pseudo-PSSM):
            o	Combines original and smoothed PSSM values.
            o	Balances local and global evolutionary signals.
    
            4.	DP-PSSM (Distance-PSSM):
            o	Computes distance-based correlations between residues.
            o	Encodes spatial dependencies in substitution patterns.
    
            5.	PSSM-AC and PSSM-CC:
            o	AC (Autocorrelation): Calculates correlations between PSSM columns for grouped residues.
            o	CC (Cross-Correlation): Calculates cross-correlations between different columns.

            #### Combined Features
            These methods concatenate multiple feature sets.
            1.	AADP-PSSM (AAC + DPC PSSM):
            o	Combines amino acid composition with di-peptide composition for a comprehensive feature set.
    
            2.	AATP (AAC + TPC):
            o	Combines amino acid composition with triplet positional information.
    
            3.	MEDP (Mean Evolutionary Difference Profile):
            o	Combines evolutionary difference features (EDP and EEDP).

            #### Key Takeaways
            	Row-based Features: Focus on sequence positions (e.g., AAC, RPM).
            	Column-based Features: Focus on substitution probabilities (e.g., DPC, TPC).
            	Mixed Features: Combine sequence and substitution information (e.g., EDP, PSE-PSSM).
            	Combined Features: Fuse multiple features for richer representations.
  Row-Based Features: Capture local and position-specific information, useful for identifying conserved regions or motifs in sequences.
  Column-Based Features: Highlight global substitution trends and relationships between residues, useful for characterizing overall sequence evolution.

# Post-Processing

```python
from Bio import SeqIO
import pandas as pd

def process_pssm_composition(filtered_fasta, pssm_csv, input_fasta, output_csv):
    """
    Process PSSM composition by merging profiles based on sequence.
    
    Parameters:
    - filtered_fasta: Path to the filtered_seq.fasta file.
    - pssm_csv: Path to the pssm_composition_protein_ntx.csv file.
    - input_fasta: Path to the cd_40_ntxprotein.fasta file.
    - output_csv: Path to save the final merged CSV file.
    """
    
    # Step 1: Convert filtered_seq.fasta to CSV
    def fasta_to_dataframe(fasta_file):
        records = []
        for record in SeqIO.parse(fasta_file, "fasta"):
            records.append({"Entry": record.id, "Sequence": str(record.seq)})
        return pd.DataFrame(records)
    
    print("Converting filtered_seq.fasta to CSV...")
    filtered_seq_df = fasta_to_dataframe(filtered_fasta)
    
    # Step 2: Add Entry and Sequence columns to pssm_composition_protein_ntx.csv
    print("Merging filtered_seq.csv with PSSM composition...")
    pssm_df = pd.read_csv(pssm_csv)
    pssm_with_sequences = pd.concat([pssm_df, filtered_seq_df], axis=1)
    
    # Step 3: Convert cd_40_ntxprotein.fasta to CSV
    print("Converting cd_40_ntxprotein.fasta to CSV...")
    input_fasta_df = fasta_to_dataframe(input_fasta)
    
    # Step 4: Merge profiles with cd_40_ntxprotein.csv based on Sequence and Entry
    print("Merging profiles with cd_40_ntxprotein.csv...")
    merged_df = pd.merge(input_fasta_df, pssm_with_sequences, on=["Sequence", "Entry"], how="left")
    
    # Step 5: Fill missing profiles with 0
    print("Filling missing profiles with 0...")
    merged_df.fillna(0, inplace=True)
    
    # Save the final DataFrame to CSV
    print(f"Saving final merged file to {output_csv}...")
    merged_df.to_csv(output_csv, index=False)
    print("Processing complete!")


# Provide paths to fasta contain filtered sequence for which pssm is generated.
filtered_fasta_path = "../neg_protein_pssm_composition/filtered_seq.fasta"
# Provide paths to specific pssm file
pssm_csv_path = "../neg_protein_pssm_composition/pssm_composition_protein_non_ntx.csv"
# Provide paths to fasta contain all sequence.
input_fasta_path = "../cd_40_non_ntxprotein2.fasta"
# Provide paths to the output file
output_csv_path = "../neg_protein_pssm_composition/cd_40_non_ntxprotein_pssm_composition.csv"

# Run the function
process_pssm_composition(filtered_fasta_path, pssm_csv_path, input_fasta_path, output_csv_path)
```
